```python
# Linear Search (O(n))
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1


# Bubble Sort (O(n^2))
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Swap
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr


# -------- Example Usage --------

numbers = [10, 100, 1000, 10000]

print("Original List:", numbers)

# Sorting the list
sorted_numbers = bubble_sort(numbers.copy())
print("Sorted List:", sorted_numbers)

# Searching for a number
target = 100
result = linear_search(sorted_numbers, target)

if result != -1:
    print(f"{target} found at index {result}")
else:
    print(f"{target} not found")
```

```python
import random
import time
import matplotlib.pyplot as plt

# -----------------------------
# Linear Search (O(n))
# -----------------------------
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1


# -----------------------------
# Bubble Sort (O(n^2))
# -----------------------------
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr


# -----------------------------
# Compare Performance
# -----------------------------

# Different dataset sizes
sizes = [10, 100, 1000, 10000]

linear_times = []
bubble_times = []

for size in sizes:
    # Generate random dataset
    data = random.sample(range(size * 10), size)
    target = data[-1]  # worst-case for linear search

    # Measure Linear Search time
    start = time.time()
    linear_search(data, target)
    end = time.time()
    linear_times.append(end - start)

    # Measure Bubble Sort time
    data_copy = data.copy()
    start = time.time()
    bubble_sort(data_copy)
    end = time.time()
    bubble_times.append(end - start)


# -----------------------------
# Plot Results
# -----------------------------

plt.figure()
plt.plot(sizes, linear_times, label="Linear Search O(n)")
plt.plot(sizes, bubble_times, label="Bubble Sort O(n^2)")

plt.xlabel("Dataset Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.title("Performance Comparison: O(n) vs O(n^2)")
plt.legend()

plt.show()
```
